```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.set_option('display.max_columns',8) # 控制台打印时显示 8 列
pd.set_option('precision', 1) # 控制台打印时显示 1 位小数

boston_data = datasets.load_boston()
df_boston = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
df_boston['target'] = pd.Series(boston_data.target)
```

## Tensorflow 模型

```python
import tensorflow as tf

W = tf.Variable(tf.truncated_normal([1], mean=0.0, stddev=1), dtype=tf.float32)
b = tf.Variable(tf.truncated_normal([1], mean=30.0, stddev=1), dtype=tf.float32)
#W = tf.Variable([1], dtype=tf.float32)
#b = tf.Variable([1], dtype=tf.float32)
x = tf.placeholder(tf.float32)

linear_model = W*x + b

y = tf.placeholder(tf.float32)

loss = tf.sqrt(tf.reduce_sum(tf.square(linear_model - y)))

#optimizer = tf.train.AdamOptimizer(0.02)
optimizer = tf.train.GradientDescentOptimizer(0.0001)
train = optimizer.minimize(loss)

x_train = df_boston.AGE
y_train = df_boston.target
train_indices = np.random.choice(len(x_train), np.int(len(x_train)*0.8), replace=False)
test_indices = np.array(list(set(range(len(x_train))) - set(train_indices)))

x_train_val = x_train[train_indices]
y_train_val = y_train[train_indices]
x_test_val = x_train[test_indices]
y_test_val = y_train[test_indices]

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
plt.scatter(x=np.array(df_boston.AGE),y=np.array(df_boston.target))
for i in range(1000):
  sess.run(train, {x: x_train_val, y: y_train_val})

  curr_W, curr_b, curr_loss_train = sess.run([W, b, loss], {x: x_train_val, y: y_train_val})
  curr_loss_test = sess.run(loss, {x: x_test_val, y: y_test_val})
  print("Train# W: %s b: %s loss: %s\n Test# loss: %s"%(curr_W, curr_b, curr_loss_train, curr_loss_test))

  xx=np.linspace(0,100,100)
  plt.plot(xx,curr_W*xx+curr_b)

plt.plot(xx,curr_W*xx+curr_b,"ro")
with tf.summary.FileWriter(logdir='logs_linear_regression', graph=tf.get_default_graph()) as writer:
    writer.flush()
```

## tf.estimator 模型

```python
from sklearn import datasets
import pandas as pd
import tensorflow as tf
from tensorflow.contrib.data import Dataset
pd.set_option('display.max_columns',8) # 控制台打印时显示 8 列
boston_data = datasets.load_boston()
df_boston = pd.DataFrame(boston_data.data,columns=boston_data.feature_names)
df_boston['target'] = pd.Series(boston_data.target)
df_boston = df_boston.reindex(np.random.permutation(df_boston.index))

df_boston.plot(kind='scatter', x='AGE', y='target',
    # c='col_for_color', s=my_df['col_for_size']*10
    );

my_feature = df_boston[["AGE"]]
targets = df_boston["target"]

feature_columns = [tf.feature_column.numeric_column("AGE")]
```

```python
#my_optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.0000001)
my_optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.0005)
my_optimizer = tf.contrib.estimator.clip_gradients_by_norm(my_optimizer, 5.0)

linear_regressor = tf.estimator.LinearRegressor(
    feature_columns=feature_columns,
    optimizer=my_optimizer
)
```

```python
def my_input_fn(features, targets, batch_size=1, shuffle=True, num_epochs=None):
    features = {key:np.array(value) for key,value in dict(features).items()}                                           
    ds = Dataset.from_tensor_slices((features,targets)) # warning: 2GB limit
    ds = ds.batch(batch_size).repeat(num_epochs)
    if shuffle:
      ds = ds.shuffle(buffer_size=10000)
    return ds.make_one_shot_iterator().get_next()
```

```python
_ = linear_regressor.train(
    input_fn = lambda:my_input_fn(my_feature, targets),
    steps=100
)
```

## keras 模型

```python
x_train = df_boston.AGE
y_train = df_boston.target
train_indices = np.random.choice(len(x_train), np.int(len(x_train)*0.8), replace=False)
test_indices = np.array(list(set(range(len(x_train))) - set(train_indices)))
x_train_val = x_train[train_indices]
y_train_val = y_train[train_indices]
x_test_val = x_train[test_indices]
y_test_val = y_train[test_indices]
```

```python
from tensorflow.python.keras.layers import Input, Dense
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.optimizers import SGD, RMSprop

model = Sequential()  
model.add(Dense(input_dim=1, units=1))  
sgd = SGD(lr=0.000001, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='mae', optimizer='sgd')  

print('Training -----------')  
for step in range(501):  
    cost = model.train_on_batch(x_train_val, y_train_val)  
    if step % 50 == 0:  
        print("After %d trainings, the cost: %f" % (step, cost))

print('\nTesting ------------')  
cost = model.evaluate(x_test_val, y_test_val, batch_size=400)  
print('test cost:', cost)  
W, b = model.layers[0].get_weights()  
print('Weights=', W, '\nbiases=', b)  

Y_pred = model.predict(x_test_val)  
plt.scatter(x_test_val, y_test_val)  
plt.plot(x_test_val, Y_pred)  
plt.show()  
```
